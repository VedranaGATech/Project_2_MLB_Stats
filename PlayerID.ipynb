```python
import pandas as pd
import pymongo
```

```python
pitching = pd.read_csv('data/pitching_final.csv')
batting = pd.read_csv('BaseballHits/Batting_Stats.csv')
pitching
```

```python
bat = batting.rename(columns={'playerid': 'playerID'})
bat
```

```python
players = pd.merge(bat, pitching, on='playerID')
```

```python
Team = players.groupby('playerID')['team'].unique()
Year = players.groupby('playerID')['year'].unique()

df = pd.DataFrame({
    "Team": Team,
    'Year': Year
})
df
```
